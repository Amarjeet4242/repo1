**question 1 (a)	Since the attributes present in this dataset is of different scales, first rescale the attributes of this dataset so that all the attributes have the same scale. Use the scale 0 to 1 for the feature information of this dataset.**

```python
# Upload dataset
from google.colab import files
uploaded = files.upload()
import pandas as pd

# Read dataset
filename = list(uploaded.keys())[0]
df = pd.read_csv(filename)

# Separate features only (exclude label)
X = df.iloc[:, :-1]   # last column is class/label

print("Original Feature Data:")
X.head()
```

```python
# Min-Max Scaling (0 to 1)
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler(feature_range=(0, 1))
X_scaled = scaler.fit_transform(X)

X_scaled_df = pd.DataFrame(X_scaled, columns=X.columns)

print("Scaled Data (0 to 1):")
X_scaled_df.head()
```

**(b)	The next is to binarize the data. Transform the whole data set using a binary threshold. Allvalues above the threshold are marked 1 and all equal to or below are marked as 0. Use thethreshold value
0.0 for all the features of this dataset. Print the head of the transformed data.**

```python
from sklearn.preprocessing import Binarizer

binarizer = Binarizer(threshold=0.0)
X_binarized = binarizer.fit_transform(X_scaled_df)

X_binarized_df = pd.DataFrame(X_binarized, columns=X.columns)

print("Binarized Data:")
X_binarized_df.head()
```

**(c)	The last step is to standardize the data, which is essential for analysing data distribution. Standardize the data associated with this dataset to a standard Gaussian distribution with amean of 0 and standard deviation of 1. (Hint. Use the scikit-learn library)**

```python
from sklearn.preprocessing import StandardScaler

standardizer = StandardScaler()
X_standardized = standardizer.fit_transform(X)

X_standardized_df = pd.DataFrame(X_standardized, columns=X.columns)

print("Standardized Data:")
X_standardized_df.head()
```

```python
from google.colab import files
uploaded = files.upload()
import pandas as pd

# Read dataset
filename = list(uploaded.keys())[0]
df = pd.read_csv(filename)

print("Original Dataset:")
df.head()
```

```python
# Separate features and label
X = df.iloc[:, :-1]   # all columns except last
Y = df.iloc[:, -1]    # last column as label

print("Features (X):")
X.head()

print("Labels (Y):")
Y.head()
```

**(b)	Make use of the label encoder to handle the categorical data present in the dataframe. the Region variable should consist of a 3-bit binary variable. The left most bit represents India,2nd bit represents Brazil and the last bit represents USA. If the bit is 1 then it represents data for that country otherwise not. For Online Shopper variable, 1 represents Yes and 0 represents No.**

```python
from sklearn.impute import SimpleImputer

# Separate numerical and categorical columns
num_cols = X.select_dtypes(include=['int64', 'float64']).columns
cat_cols = X.select_dtypes(include=['object']).columns

# Mean imputation for numerical data
num_imputer = SimpleImputer(strategy='mean')
X[num_cols] = num_imputer.fit_transform(X[num_cols])

# Most frequent imputation for categorical data
cat_imputer = SimpleImputer(strategy='most_frequent')
X[cat_cols] = cat_imputer.fit_transform(X[cat_cols])

print("Dataset after Imputation:")
X.head()
```

**(c) Use the StandardScaler() to scale the dataset, where all the attributes are of same sacale.**

```python
import pandas as pd

# Remove extra spaces from column names (VERY IMPORTANT)
X.columns = X.columns.str.strip()
print("Columns:", X.columns)

# If column name is 'Online Shopper'
if 'Online Shopper' in X.columns:
    X['Online Shopper'] = X['Online Shopper'].map({'Yes': 1, 'No': 0})

elif 'Online_Shopper' in X.columns:
    X['Online_Shopper'] = X['Online_Shopper'].map({'Yes': 1, 'No': 0})
region_encoded = pd.get_dummies(X['Region'])

region_encoded = region_encoded[['India', 'Brazil', 'USA']]

X.drop(columns=['Region'], inplace=True)


X = pd.concat([X, region_encoded], axis=1)

print("Encoded Dataset (Part c Output):")
X.head()

```
